**<center> PRUEBA DE KOLMOGOROV-SMIRNOV</center>**


En esta prueba ordenamos la muestra $X_n$ bajo examen de menor a mayor y comparamos la **frecuencia observada** contra la **frecuencia esperada**; es decir compararemos contra la distribuciòn acumulativa, F(x).

**PASOS A SEGUIR**

Dado una muestra $X_n$ de prueba.

Establecemos:

$H_0: X_n$ $ \sim U(0, 1)$ Establecemos $\alpha$

$H_1: X_n$ $ \sim U(0, 1)$

**PASO 1:**

Ordenar los nùmeros de la muestra en orden ascendente.


**PASO 2:**


Calcular el estadìstico de prueba a partir de:

$D^{+}$ = max$_{1 < i < n}$ $\big | \frac {i}{n} -X_i\big |$

$D^{-}$ = max$_{1 < i < n}$ $\big |X_i- \frac {i}{n} \big |$

**D**= max ($D^{+}, D^{-}$) i es la posición de $X^{i}$

**PASO 3:**

Comparar con el valor más crítico:

Si D < d$\alpha$ no se rechaza $H_0$

En la tabla lo nombremos como $D^{-}$ ó también comonocido como **D a la menos.**

```python
#Recordamos quien es n ó en caso de que no esté, definirlo y llamar a X_i
n = 100
for i in range(1, n + 1):
    D_menos = abs(X_i[i-1] - ((i - 1) / n))
    print(D_menos)
```

Y por último en la tabla lo nombremos como $D^{+}$ ó también comonocido como **D a la positiva**

```python
# Tenemos 100 valores ó numbres o cantidades
n = 100

# Calculamos e imprimimos D+
for i in range(1, n + 1):
    # Obtenemos el dato actual de tu lista X_i
    dato_actual = X_i[i-1]

    # Aplicamos la fórmula para D+ : valor absoluto de [(i/n) - Xi]
    D_mas = abs((i / n) - dato_actual)

    print(D_mas)
```

Ahora con todos los datos obtenidos con anterioirdad, creamos una gráfica que tntenga los fadts anteiores.

```python
import pandas as pd
import numpy as np

# Definimos n basándonos en la longitud
n = len(X_i)
i = np.arange(1, n + 1)

# Creamos la tabla la variable X_i
tabla = pd.DataFrame({
    'i': i,
    'X_i': X_i,
    '(i-1)/n': (i - 1) / n,
    'D-': np.abs(np.array(X_i) - ((i - 1) / n)),
    'D+': np.abs((i / n) - np.array(X_i))
})

# Imprimimos la tabla completa (las 100 filas)
print(tabla.to_string(index=False))
```
